```python
# PROJECT 1: Simple Interest Calculator

#flowchart
![Simple Interest Flowchart](flowchart1.png)
print("SIMPLE INTEREST CALCULATOR")

# Get input from user
P = float(input("Enter Principal amount (P): "))
R = float(input("Enter Interest rate (R): "))
T = float(input("Enter Time period (T): "))

# Calculate Simple Interest
interest = (P * R * T) / 100
A = P + interest

# Display results

print("RESULTS")
print(f"Principal Amount: {P:,.2f}")
print(f"Interest Rate: {R}%")
print(f"Time : {T} years")
print(f"Interest Earned: {interest:,.2f}")
print(f"Final Amount: {A:,.2f}")
```
