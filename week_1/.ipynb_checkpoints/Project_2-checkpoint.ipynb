#Compound Interest Calculator
# Flowchart
![Compound Interest Flowchart](flowchart2.png)

print("COMPOUND INTEREST CALCULATOR")

# Get input from user
print("\nPlease enter the following values:")
P = float(input("Principal amount (P): "))
R = float(input("Annual interest rate (R) in %: "))
t = float(input("Time period (t) in years: "))
n = int(input("Number of times interest compounded per year (n): "))


# Calculate Compound Interest
r = R / 100  # Convert percentage to decimal
A = P * (1 + r/n) ** (n * t)

# Calculate the interest earned
interest_earned = A - P

print("RESULTS")
print(f"Principal Amount: ${P:,.2f}")
print(f"Annual Interest Rate: {R}%")
print(f"Time Period: {t} years")
print(f"Compounding Frequency: {n} time(s) per year")


print(f"Final Amount: {A:,.2f}")
print(f"Interest Earned: {interest_earned:,.2f}")
