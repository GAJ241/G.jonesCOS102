print("=" * 60)
print("This calculates the future value of a series of regular payments")
print("=" * 60)

# Get input from user
print("\nPlease enter the following values:")
PMT = float(input("Payment amount per period (PMT): $"))
R = float(input("Annual interest rate (R) in %: "))
t = float(input("Time period (t) in years: "))
n = int(input("Number of payments per year (n): "))


# Calculate Annuity Future Value
r = R / 100  # Convert percentage to decimal
rate_per_period = r / n
number_of_payments = n * t

# Future value of annuity
A = PMT * ((1 + rate_per_period) ** number_of_payments - 1) / rate_per_period

# Total payments made
total_payments = PMT * number_of_payments

# Total interest earned
total_interest = A - total_payments

# Display results
print(f"Payment per period: {PMT:,.2f}")
print(f"Annual interest rate: {R}%")
print(f"Time period: {t} years")
print(f"Payments per year: {n}")
print(f"Total number of payments: {int(number_of_payments)}")
print(f"Future Value of Annuity: ${A:,.2f}")
print(f"Total payments made: ${total_payments:,.2f}")
print(f"Total interest earned: ${total_interest:,.2f}")
