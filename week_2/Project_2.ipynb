```python
print("Equation Solver")
print("1. Quadratic (Ax^2 + Bx + C = 0)")
print("2. Cubic (Ax^3 + Bx^2 + Cx + D = 0)")
print("3. Quartic (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")

choice = int(input("Choose the equation type: "))

# QUADRATIC
if choice == 1:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))

    d = b*b - 4*a*c

    x1 = (-b + d**0.5) / (2*a)
    x2 = (-b - d**0.5) / (2*a)

    print("Root 1 =", x1)
    print("Root 2 =", x2)

# CUBIC
elif choice == 2:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))

    x = float(input("Enter a value for x to test: "))

    value = a*x**3 + b*x**2 + c*x + d

    if value == 0:
        print("x =", x, "is a root")
    else:
        print("x =", x, "is not a root")

# QUARTIC
elif choice == 3:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    x = float(input("Enter a value for x to test: "))

    value = a*x**4 + b*x**3 + c*x**2 + d*x + e

    if value == 0:
        print("x =", x, "is a root")
    else:
        print("x =", x, "is not a root")

else:
    print("Invalid choice")
```
